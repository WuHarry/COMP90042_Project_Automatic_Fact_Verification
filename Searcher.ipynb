```python
INDEX_DIR = "IndexFiles.index"

import Ipynb_importer
import Indexing

import sys, os, lucene, time
from datetime import datetime
from java.nio.file import Paths
from org.apache.lucene.analysis.standard import StandardAnalyzer
from org.apache.lucene.index import DirectoryReader
from org.apache.pylucene.queryparser.classic import PythonMultiFieldQueryParser
from org.apache.lucene.queryparser.classic import QueryParser
from org.apache.lucene.store import SimpleFSDirectory
from org.apache.lucene.search import IndexSearcher, BooleanClause

def run(searcher, analyzer):
    while True:
        print()
        print("Hit enter with no input to quit.")
        command = input("Query:")
        if command == '':
            return

        print()
        print("Searching for:", command)
          
        query = PythonMultiFieldQueryParser(['name', 'contents'], analyzer).parse(command, ['name', 'contents'],
                         [BooleanClause.Occur.SHOULD, BooleanClause.Occur.SHOULD], analyzer)
        scoreDocs = searcher.search(query, 50).scoreDocs
        print("%s total matching documents." % len(scoreDocs))

        for scoreDoc in scoreDocs:
            doc = searcher.doc(scoreDoc.doc)
            print('docname:', doc.get("docname"), 'name:', doc.get("name"), 'content:', doc.get("contents"))


if __name__ == '__main__':
    assert lucene.getVMEnv() or lucene.initVM(vmargs=['-Djava.awt.headless=true'])
    print('lucene', lucene.VERSION)
#     start = datetime.now()
#     try:
#         base_dir = os.getcwd()
#         Indexing.IndexFiles('wiki-pages-text.zip', os.path.join(base_dir, INDEX_DIR), StandardAnalyzer())
#         end = datetime.now()
#         print(end - start)
#     except Exception as e:
#         print("Failed: ", str(e))
        
    base_dir = os.getcwd()
    directory = SimpleFSDirectory(Paths.get(os.path.join(base_dir, INDEX_DIR)))
    searcher = IndexSearcher(DirectoryReader.open(directory))
    analyzer = StandardAnalyzer()
    run(searcher, analyzer)
    del searcher
```
